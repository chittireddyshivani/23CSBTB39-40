```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Data
data = {
    "Daily Calorie Intake": [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000],
    "Exercise Duration": [60, 50, 30, 90, 45, 20, 75, 95, 15, 10],
    "BMI": [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]
}

df = pd.DataFrame(data)

# Scatter plots
plt.figure(figsize=(15, 5))

# Daily Calorie Intake vs. BMI
plt.subplot(1, 3, 1)
plt.scatter(df["Daily Calorie Intake"], df["BMI"], color='b')
plt.xlabel("Daily Calorie Intake")
plt.ylabel("BMI")
plt.title("Daily Calorie Intake vs. BMI")

# Exercise Duration vs. BMI
plt.subplot(1, 3, 2)
plt.scatter(df["Exercise Duration"], df["BMI"], color='r')
plt.xlabel("Exercise Duration")
plt.ylabel("BMI")
plt.title("Exercise Duration vs. BMI")

# Daily Calorie Intake vs. Exercise Duration
plt.subplot(1, 3, 3)
plt.scatter(df["Daily Calorie Intake"], df["Exercise Duration"], color='g')
plt.xlabel("Daily Calorie Intake")
plt.ylabel("Exercise Duration")
plt.title("Daily Calorie Intake vs. Exercise Duration")

plt.tight_layout()
plt.show()

# Correlation Coefficients
pairs = [("Daily Calorie Intake", "BMI"),
         ("Exercise Duration", "BMI"),
         ("Daily Calorie Intake", "Exercise Duration")]

for x, y in pairs:
    corr, _ = pearsonr(df[x], df[y])
    print(f"Pearson correlation between {x} and {y}: {corr:.2f}")
```
